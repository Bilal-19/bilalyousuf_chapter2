```python
import sqlite3
from sqlite3 import Error
import tkinter
from tkinter import *


con=sqlite3.connect('Book.db')
cursorObj=con.cursor()
cursorobj.execute("INSERT INTO Table1 VALUES(1,'Calculus','Aanton','howard',2019,1000,350,'Yes')")
cursorobj.execute("INSERT INTO Table1 VALUES(2,'English','Henry','Selvister',2017,800,250,'Yes')")
cursorobj.execute("INSERT INTO Table1 VALUES(3,'ICT','George','David',2016,900,350,'Yes')")
cursorobj.execute("INSERT INTO Table1 VALUES(4,'Progarmming Fundamental','Alis','Bag',206,1500,400,'Yes')")
cursorobj.execute("INSERT INTO Table1 VALUES(5,'Basic Electronics','Mark','Davis',2018,1000,180,'No')")
con.commit()

window=Tk()
window.title("BOOKS")
window.geometry("800x300")
window.resizable(0,0)

headLbl=Label(window,text="DIFFERENT QUERIES",font=("Arial",30,"bold"))
headLbl.pack(pady=10)
selectQuery=StringVar()
updateQuery=StringVar()
selectBookQuery=StringVar()
softAvailQuery=StringVar()


frame=Frame(window)
frame.pack()
lbl1=Label(frame,text="Select wrt ID",font=10,relief='sunken',width=20)
lbl1.grid(row=0,column=0)
cursorObj.execute("Select bookTitle from Table1 where BookId='5'")
query1=cursorObj.fetchall()
selectQuery.set(query1)
lbl2=Label(frame,textvariable=selectQuery,font=8)
lbl2.grid(row=0, column=1)

lbl=Label(frame,text="Update",font=10,relief='sunken',width=20)
lbl.grid(row=2,column=0)
cursorObj.execute("UPDATE Table1 SET bookAuthor='Perkovic' where BookId=3")
cursorObj.execute("Select * FROM Table1 where BookId=3")
query2=cursorObj.fetchall()
updateQuery.set(query2)
lbl3=Label(frame,textvariable=updateQuery,font=8)
lbl3.grid(row=2, column=1)

lbl=Label(frame,text="Select All",font=10,relief='sunken',width=20)
lbl.grid(row=3,column=0)
cursorObj.execute("Select * from Table1 where bookTitle='Calculus' ")
query3=cursorObj.fetchall()
selectBookQuery.set(query3)
lbl4=Label(frame,textvariable=selectBookQuery,font=8)
lbl4.grid(row=3, column=1)

lbl=Label(frame,text="Availble Books",font=10,relief='sunken',width=20)
lbl.grid(row=4,column=0)
cursorObj.execute("SELECT bookTitle from Table1 where softcopyavailable='Yes'")
query4=cursorObj.fetchall()
softAvailQuery.set(query4)
lbl5=Label(frame,font=8,textvariable=softAvailQuery)
lbl5.grid(row=4, column=1)

window.mainloop()
```
